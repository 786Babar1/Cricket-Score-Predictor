```python
eligible_cities = df['city'].value_counts()[df['city'].value_counts() > 600].index.tolist()
```

```python
eligible_cities
```

```python
df = df[df['city'].isin(eligible_cities)]
```

```python
df.head()
```

```python
df.shape
```

```python
df['current_score'] = df.groupby('match_id')['runs'].cumsum()
```

```python
df.head()
```

```python
df['over'] = df['ball'].apply(lambda y: str(y).split(".")[0])
```

```python
df['ball_no'] = df['ball'].apply(lambda y: str(y).split(".")[1])
```

```python
df.head()
```

```python
df.tail()
```

```python
df['ball_bowled'] = (df['over'].astype(int) * 6 + df['ball_no'].astype(int))
```

```python
df.tail()
```

```python
df['ball_left'] = 120 - df['ball_bowled']
```

```python
df.head()
```

```python
df.tail()
```

```python
df['ball_left'] = df['ball_left'].apply(lambda x: 0 if x < 0 else x)
```

```python
df.tail()
```

```python
df['player_dismissed'] = df['player_dismissed'].apply(lambda x: 1 if x != '0' else '0')
```

```python
df.tail()
```

```python
df['player_dismissed'] = df['player_dismissed'].astype(int)
```

```python
res = df.dtypes
print(res)
```

```python
df['player_dismissed'] = df.groupby('match_id')['player_dismissed'].cumsum()
```

```python
df.tail()
```

```python
df['wicket_left'] = 10 - df['player_dismissed']
```

```python
df.head()
```

```python
df['current_run_rate'] = (df['current_score'] * 6) / df['ball_bowled']
```

```python
df.head()
```

```python
df.tail()
```

```python
groups = df.groupby('match_id')
#5 overs = 30 balls
match_id = df['match_id'].unique()
last_five = []

```

```python
for id in match_id:
    last_five.extend(groups.get_group(id).rolling(window=30)['runs'].sum().values.tolist())
```

```python
df['last_five'] = last_five
```

```python
df.tail()
```

```python
df.tail(18)
```

```python
final_df = df.groupby('match_id').sum()['runs'].reset_index().merge(df, on = 'match_id')
```

```python
final_df
```

```python
final_df.columns
```

```python
final_df = final_df[['batting_team', 'bowling_team', 'city', 'current_score', 'ball_left', 'wicket_left','current_run_rate', 'last_five', 'runs_x']]
```

```python
final_df.dropna(inplace = True)
```

```python
final_df
```

```python
final_df.isnull().sum()
```

```python
final_df.shape
```

```python
final_df = final_df.sample(final_df.shape[0])
```

```python
final_df.shape
```

```python
x = final_df.drop(columns = ['runs_x'])
y = final_df['runs_x']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42)
```

```python
x_train
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from xgboost import XGBRegressor
#from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score, mean_absolute_error
```

```python
blue = ColumnTransformer([('blue', OneHotEncoder(sparse_output = False, drop = 'first'), 
['batting_team', 'bowling_team', 'city'])],
 remainder = 'passthrough')
```

```python
mypipe = Pipeline(steps = [
    ('step no 1', blue),
    ('step no 2', StandardScaler()),
    ('step no 3', XGBRegressor(n_estimators = 1000, learning_rate = 0.2, max_depth = 12, random_state = 1))
])
```

```python
mypipe.fit(x_train, y_train)
y_predict_kro = mypipe.predict(x_test)
print('r2_score is: ')
print(r2_score(y_test, y_predict_kro))
print('mean_absolute_error is: ')
print(mean_absolute_error(y_test, y_predict_kro))
```

```python
import pickle
pickle.dump(mypipe, open('ready_pipe.pkl', 'wb'))
```
